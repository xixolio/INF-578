## c)

Como se puede observar a partir de la descripci√≥n de los atributos, √©stos toman valores en escalas muy diferentes, lo cual si bien no influye directamente a la regresi√≥n lineal b√°sica (debido a que el √≥ptimo se encuentra de forma anal√≠tica), cuando se optimice para lasso u otros algoritmos que utilicen optimizaci√≥n basada en gradiente descendente, la influencia de ciertos par√°metros ser√° mayor que la de otros sobre el error cuadr√°tico medio.

```python
X=df
y = np.log(df['price'])
mask = np.zeros(len(X))
limit = int(len(X)*0.7)
mask[:limit] = 1
istrain = (mask== 1)
Xtrain = X[istrain]
ytrain = y[istrain]
Xtest = X[np.logical_not(istrain)]
ytest = y[np.logical_not(istrain)]
Xtrain = Xtrain.drop(['price'], axis=1)
Xtest = Xtest.drop(['price'], axis=1)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(Xtrain)
df_scaled_train = pd.DataFrame(scaler.transform(Xtrain), columns=Xtrain.columns)
df_scaled_train.insert(df_scaled_train.shape[1], 'intercept', np.ones(df_scaled_train.shape[0]))
```

## d) 

Los argumentos para la regresi√≥n lineal el conjunto de datos X, que contiene las muestras, y el conjunto de etiquetas Y. El par√°metro fit_intercept define la inclusi√≥n o no del sesgo (par√°metro $ B_0$), por lo que, debido a que agregamos una columna de 1's a nuestro conjunto de datos X, no es necesario.

```python
import sklearn.linear_model as lm
linreg = lm.LinearRegression(fit_intercept = False)
linreg.fit(df_scaled_train, ytrain)
mse = np.mean(linreg.predict(df_scaled_train) - ytrain)
```

## e)

Estimaci√≥n de la varianza del error:

```python
N = limit
I = len(df_scaled_train.columns)
sigma = np.sum((linreg.predict(df_scaled_train) - ytrain)**2)/( N - I - 1 )
#sigma2 = np.dot((linreg.predict(df_scaled_train) - ytrain).transpose(),(linreg.predict(df_scaled_train) - ytrain))/(N-I-1)
```

Estimaci√≥n de la matriz de covarianzas de los par√°metros:

```python
cov_matrix = np.dot(df_scaled_train.values.transpose(),df_scaled_train.values)
np.linalg.cond(cov_matrix)
#inv_cov_matrix = np.linalg.inv(cov_matrix)
#u,v = np.linalg.eig(inv_cov_matrix)
```

Notamos que el n√∫mero de condici√≥n de la matrix de covarianza es muy alto, lo cual nos indica que las columnas/filas est√°n cercanas a ser linealmente dependientes. Se buscar√° la relaci√≥n de dependencia lineal utilizando la correlaci√≥n entre las caracter√≠sticas. 

```python
df_scaled_train.corr()
```

Es notoria la correlaci√≥n entre las variables sqft_living, sqft_basement y sqft_above. Seg√∫n la descripci√≥n de las variables otorgada previamente, se puede deducir que sqft_living = sqft_basement + sqft_above. Si √©sto fuera as√≠, la matriz no ser√≠a invertible. Sin embargo lo es, pero √©sto puede ser producto de haber ingresado mal los datos o de errores de medici√≥n y c√°lculo. Veremos si los z-score son capaces de dar cuenta de esta relaci√≥n ofreci√©ndonos eliminar alguna de estas variables.

```python
linreg = lm.LinearRegression(fit_intercept = False)
linreg.fit(df_scaled_train, ytrain)
I = len(df_scaled_train.columns)
B = linreg.coef_
sigma = np.sum((linreg.predict(df_scaled_train) - ytrain)**2)/(limit - I - 1 )
stds = np.sqrt(np.diag(np.linalg.inv(cov_matrix))*sigma)
z_scores = B/stds
print(z_scores)
```

```python
from scipy.stats import t
p_value = t.ppf(0.975,df=limit - I - 1)
np.abs(z_scores) > p_value
print(df_scaled_train.columns)
print(df_scaled_train.columns[np.abs(z_scores) > p_value])
print(df_scaled_train.columns[np.abs(z_scores) <= p_value])
```

La √∫ltima lista nos muestra las variables candidatas a ser eliminadas. Efectivamente, el z-score es caapz de lidiar con el problema de las variables altamente correlacionadas.

```python
df_scaled_train = df_scaled_train.drop(['sqft_living'], axis=1)
cov_matrix = np.dot(df_scaled_train.values.transpose(),df_scaled_train.values)
print("nuevo n√∫mero de condici√≥n " + str(np.linalg.cond(cov_matrix)))
linreg = lm.LinearRegression(fit_intercept = False)
linreg.fit(df_scaled_train, ytrain)
I = len(df_scaled_train.columns)
B = linreg.coef_
sigma = np.sum((linreg.predict(df_scaled_train) - ytrain)**2)/(limit - I - 1 )
stds = np.sqrt(np.diag(np.linalg.inv(cov_matrix))*sigma)
z_scores = B/stds
print(z_scores)
```

El n√∫mero de condici√≥n de la matriz disminuye notoriamente. La matriz ahora s√≠ est√° bien condicionada y sus valores propios ya no deben ser cercanos a 0.

```python
p_value = t.ppf(0.975,df=limit - I - 1)
np.abs(z_scores) > p_value
print(df_scaled_train.columns)
print(df_scaled_train.columns[np.abs(z_scores) > p_value])
print(df_scaled_train.columns[np.abs(z_scores) <= p_value])
```

Construyendo nuevamente el modelo, notamos que la √∫nica variable cuyo z-score es menor que el p-value asociado es long.

```python
df_scaled_train = df_scaled_train.drop(['long'], axis=1)
cov_matrix = np.dot(df_scaled_train.values.transpose(),df_scaled_train.values)
linreg = lm.LinearRegression(fit_intercept = False)
linreg.fit(df_scaled_train, ytrain)
I = len(df_scaled_train.columns)
B = linreg.coef_
sigma = np.sum((linreg.predict(df_scaled_train) - ytrain)**2)/(limit - I - 1 )
stds = np.sqrt(np.diag(np.linalg.inv(cov_matrix))*sigma)
z_scores = B/stds
print(z_scores)
```

```python
p_value = t.ppf(0.975,df=limit - I - 1)
np.abs(z_scores) > p_value
print(df_scaled_train.columns)
print(df_scaled_train.columns[np.abs(z_scores) > p_value])
print(df_scaled_train.columns[np.abs(z_scores) <= p_value])
```

Finalmente, todas las variables superan el test y ninguna es eliminada.

# 2 Selecci√≥n de Atributos

Comenzamos todo de 0 en este punto de manera de tener los datos de entrada *limpios*.

```python
X=df
y = np.log(df['price'])
mask = np.zeros(len(X))
limit = int(len(X)*0.7)
mask[:limit] = 1
istrain = (mask== 1)
Xtrain = X[istrain]
ytrain = y[istrain]
Xtest = X[np.logical_not(istrain)]
ytest = y[np.logical_not(istrain)]
Xtrain = Xtrain.drop(['price'], axis=1)
Xtest = Xtest.drop(['price'], axis=1)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(Xtrain)
df_scaled_train = pd.DataFrame(scaler.transform(Xtrain), columns=Xtrain.columns)
df_scaled_train.insert(df_scaled_train.shape[1], 'intercept', np.ones(df_scaled_train.shape[0]))
df_scaled_test = pd.DataFrame(scaler.transform(Xtest), columns=Xtrain.columns)
df_scaled_test.insert(df_scaled_test.shape[1], 'intercept', np.ones(df_scaled_test.shape[0]))
```

```python
def fss(df_x, df_y, names_x, k = 10000):
    x = df_x.as_matrix()
    y = df_y.as_matrix()
    columns = df_x.columns
    p = x.shape[1]
    k = min(p, k)
    names_x = np.array(names_x)
    remaining = list(range(0, p))
    selected = []
    current_score = 0.0
    best_new_score = 0.0
    while remaining and len(selected) <= k:
        print remaining
        score_candidates = []
        kf = KFold(n_splits=10)
        for candidate in remaining:
            model = lm.LinearRegression(fit_intercept=False)
            indexes = selected + [candidate]
            x_t = x[:,indexes]
                        
            mse_cv_candidate = 0
            for train, val in kf.split(x_t):
                cross_scaler = StandardScaler().fit(x_t[train])
                df_cross_train = pd.DataFrame(cross_scaler.transform(x_t[train]),columns=columns[indexes])
                df_cross_train.insert(df_cross_train.shape[1], 'intercept', np.ones(df_cross_train.shape[0]))
                linreg = lm.LinearRegression(fit_intercept = False)
                linreg.fit(df_cross_train, y[train])
                df_cross_test = pd.DataFrame(cross_scaler.transform(x_t[val]),columns=columns[indexes])
                df_cross_test.insert(df_cross_test.shape[1], 'intercept', np.ones(df_cross_test.shape[0]))
                yhat_val = linreg.predict(df_cross_test)
                mse_fold = np.mean(np.power(yhat_val - ym[val], 2))
                mse_cv_candidate += mse_fold
                      
            score_candidates.append((mse_cv_candidate, candidate))
            
        score_candidates.sort()
        score_candidates[:] = score_candidates[::-1]
        best_new_score, best_candidate = score_candidates.pop()
        remaining.remove(best_candidate)
        selected.append(best_candidate)
        
        print("selected = %s ..." % names_x[best_candidate])
        print("totalvars=%d, mse = %f" % (len(indexes), best_new_score))
    return selected

'''Revise si la cantidad de etiquetas concuerda con la cantidad de variables que esta utilizando,
puede que tenga que eliminar un elemento de la lista name_regressors'''

names_regressors = [
    "bedrooms",
    "bathrooms",
    "sqft_living",
    "sqft_lot",
    "floors",
    "waterfront",
    "view",
    "condition",
    "grade",
    "sqft_above",
    "sqft_basement",
    "yr_built",
    "yr_renovated",
    "lat-long",
    "long",
    "sqft_living15",
    "sqft_lot15"]


x = df_scaled_train.drop(['intercept'],axis=1)
y = ytrain

selected = fss(x,y,names_regressors)
```

El cambio realizado al criterio de selecci√≥n de variables consiste en realizar cross-validation para cada candidato, y utilizar el error cuadr√°tico promedio del cross para determinar qu√© variable es la que nos produce un modelo con mejor capacidad de generalizaci√≥n. Si bien √©sto no es tan √∫til debido a que el train set y test set aparentemente se distribuyen de manera similar y el supuesto de normalidad de los errores es v√°lido para ambos (lo cual implica que el mejor modelo es uno lineal), en casos en donde la distribuci√≥n var√≠e sustancialmente este criterio nos permite evitar hasta cierto punto el overfitting. Notar que despu√©s de que cada candidato ha sido seleccionado, los folds cambian para la siguiente ronda, para as√≠ evitar overfitting sobre los folds de validaci√≥n seleccionados.

```python
#columns = df_scaled_train.columns
p = df_scaled_train.shape[1]-1
index = [p]
x = df_scaled_train.as_matrix()
x_ts = df_scaled_test.as_matrix()
y = ytrain.as_matrix()
mses_tr = []
mses_ts = []
for s in selected:
    index = [s] + index
    x_s=x[:,index]
    x_ts_s = x_ts[:,index]    
    linreg = lm.LinearRegression(fit_intercept = False)
    linreg.fit(x_s, y)
    mse_tr = np.mean((y - linreg.predict(x_s))**2)
    mse_ts = np.mean((ytest.values - linreg.predict(x_ts_s))**2)
    mses_tr.append(mse_tr)
    mses_ts.append(mse_ts)

x_axis = np.arange(0,p,1)
fig = plt.figure()
plt.plot(x_axis,mses_tr,c='g',label='mse train')
plt.plot(x_axis,mses_ts,c='b',label='mse test')
plt.legend()
plt.show()
    
```

A partir de los gr√°ficos hay dos cosas importantes a considerar:

-Existe cierto grado de overfitting inicialmente al considerar tan s√≥lo unas cuantas variables. √âsto se manifiesta en el hecho de que el error sobre el train es mayor que el error sobre el test, y probablemente se debe a que, debido a que se cuenta con tan pocas variables para ajustar el modelo, los par√°metros del modelo se seleccionan buscando compensar por el error inherente a la no contribuci√≥n de las variables restantes, lo cual produce overfitting.

-Despu√©s de cierto punto la disminuci√≥n del error es insignificante. √âsto puede deberse a dos razones: existen variables que realmente no aportan casi nada al modelo (tales como long, seg√∫n el an√°lisis de los z-score), y variables que al estar muy correlacionadas, su contribuci√≥n al modelo es pr√°cticamente id√©ntica en conjunto. Por ejemplo, al agregar sqft_living casi al final, √©sta no contribuye en pr√°cticamente nada debido a que ya han sido agregadas sqft_above y sqft_basement.

# 3 Regularizaci√≥n

```python
import pandas as pd
import numpy as np
import sklearn.linear_model as lm
df = pd.read_csv("kc_house_data.csv", sep = ",", header = 0)
df = df.drop(['id', 'date', 'zipcode'], axis=1)

X = df
y = np.log(df['price'])
mask = np.zeros(len(X))
limit = int(len(X) * 0.7)
mask[:limit] = 1
istrain = (mask == 1)
Xtrain = X[istrain]
ytrain = y[istrain]
Xtest = X[np.logical_not(istrain)]
ytest = y[np.logical_not(istrain)]
Xtrain = Xtrain.drop(['price'], axis=1)
Xtest = Xtest.drop(['price'], axis=1)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(Xtrain)
df_scaled_train = pd.DataFrame(scaler.transform(Xtrain), columns=Xtrain.columns)
df_scaled_train.insert(df_scaled_train.shape[1], 'intercept', np.ones(df_scaled_train.shape[0]))
```

```python
from sklearn.linear_model import Ridge
import matplotlib.pylab as plt
Dtrain = df_scaled_train.copy()
Dtrain = Dtrain.drop('intercept', axis=1)
names_regressors = [
    "bedrooms",
    "bathrooms",
    "sqft_living",
    "sqft_lot",
    "floors",
    "waterfront",
    "view",
    "condition",
    "grade",
    "sqft_above",
    "sqft_basement",
    "yr_built",
    "yr_renovated",
    "lat-long",
    "sqft_living15",
    "sqft_lot15"]
alphas_ = np.logspace(5, -1, base=10)
coefs = []
model = Ridge(fit_intercept=True, solver='svd')
for a in alphas_:
    model.set_params(alpha=a)
    model.fit(Dtrain, ytrain)
    coefs.append(model.coef_)
ax = plt.gca()
for y_arr, label in zip(np.squeeze(coefs).T, names_regressors):
    plt.plot(alphas_, y_arr, label=label)
plt.legend()
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Regularization Path RIDGE')
plt.axis('tight')
plt.legend(loc=1)
plt.show()
```

Observamos que efectivamente, a mayor par√°metro de regularizaci√≥n $\alpha$, los coeficientes disminuyen, ya que afectan m√°s en la funci√≥n objetivo a minimizar. √âstos empiezan a converger gradualmente a 0 a medida que el alpha crece, penalizando arduamente a aquellos par√°metros que tienen una norma muy grande inicialmente ya que √©stos aumentan en demas√≠a la funci√≥n objetivo.

El intercepto no se agrega como par√°metro debido a que si lo hici√©ramos, dado un lambda grande, estar√≠amos forzando a que el hiperplano pase cerca del origen, lo cual no es necesariamente cierto.

```python
from sklearn.linear_model import Lasso
import matplotlib.pylab as plt
Dtrain = df_scaled_train.copy()
Dtrain = Dtrain.drop('intercept', axis=1)
names_regressors = [
    "bedrooms",
    "bathrooms",
    "sqft_living",
    "sqft_lot",
    "floors",
    "waterfront",
    "view",
    "condition",
    "grade",
    "sqft_above",
    "sqft_basement",
    "yr_built",
    "yr_renovated",
    "lat-long",
    "sqft_living15",
    "sqft_lot15"]
alphas_ = np.logspace(0, -3, base=10)
coefs = []
for a in alphas_:
    model = Lasso(fit_intercept=True)
    model.set_params(alpha=a)
    model.fit(Dtrain, ytrain)
    coefs.append(model.coef_)
ax = plt.gca()
for y_arr, label in zip(np.squeeze(coefs).T, names_regressors):
    plt.plot(alphas_, y_arr, label=label)
plt.legend()
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Regularization Path LASSO')
plt.axis('tight')
plt.legend(loc=1)
plt.show()
```

Lasso es efectivo para seleccionar atributos, ya que lleva algunos coeficientes efectivamente a 0. Esto ocurre por la forma de la funci√≥n de costos. Se forma un *cuadrado* alrededor del or√≠gen, a diferencia del c√≠rculo de Ridge, haciendo que sea m√°s probable que los m√≠nimos de la funci√≥n objetivo se logren en las esquinas, donde algunos coeficientes son 0. De hecho, cuando el valor del par√°metro es demasiado alto, el mejor ajuste es cuando todos los coeficientes son 0, dejando una funci√≥n constante.

```python
Dtrain = df_scaled_train.copy()
Dtrain = Dtrain.drop(['intercept'], axis=1)

#Dtrain.drop(["sqft_living"], axis=1)
#Xtest.drop(["sqft_living"], axis=1)
# Es necesario escalar los datos de entrada de test para usar un modelo
# ajustado usando valores ya escalados
df_scaled_test = pd.DataFrame(scaler.transform(Xtest), columns=Xtest.columns)

model = Ridge(fit_intercept=True)
mse_test = []
mse_train = []

alphas_ = np.logspace(3.6, 1, base=10)
for a in alphas_:
    model.set_params(alpha=a)
    model.fit(Dtrain, ytrain)
    yhat_train = model.predict(Dtrain)
    yhat_test = model.predict(df_scaled_test)
    mse_train.append(np.mean(np.power(yhat_train - ytrain, 2)))
    mse_test.append(np.mean(np.power(yhat_test - ytest, 2)))
ax = plt.gca()
ax.plot(alphas_, mse_train, label='train error ridge')
ax.plot(alphas_, mse_test, label='test error ridge')
plt.legend(loc=2)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])
plt.show()
```

Existe un rango donde la regularizaci√≥n Ridge mejora el error de test, en comparaci√≥n con no usar regularizaci√≥n (cuando $\alpha \rightarrow 0$. Este rango no coincide con donde se minimiza el error de entrenamiento. Esto √∫ltimo sucede cuando no hay regularizaci√≥n, ya que el modelo puede ajustarse aun m√°s a los datos. Cuando el par√°metro es demasiado grande, entonces la regularizaci√≥n pasa a ser m√°s importante que los datos de entrenamiento, y el error aumenta tanto para el entrenamiento como para el test.

```python
Dtrain = df_scaled_train.copy()
Dtrain = Dtrain.drop('intercept', axis=1)
# Es necesario escalar los datos de entrada de test para usar un modelo
# ajustado usando valores ya escalados
df_scaled_test = pd.DataFrame(scaler.transform(Xtest), columns=Xtest.columns)

mse_test = []
mse_train = []

alphas_ = np.logspace(0.1, -2.5, base=10)
for a in alphas_:
    model = Lasso(fit_intercept=True)
    model.set_params(alpha=a)
    model.fit(Dtrain, ytrain)
    yhat_train = model.predict(Dtrain)
    yhat_test = model.predict(df_scaled_test)
    mse_train.append(np.mean(np.power(yhat_train - ytrain, 2)))
    mse_test.append(np.mean(np.power(yhat_test - ytest, 2)))
ax = plt.gca()
ax.plot(alphas_, mse_train, label='train error ridge')
ax.plot(alphas_, mse_test, label='test error ridge')
plt.legend(loc=3)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])
plt.show()
```

En este caso particular, utilizar Lasso no mejora notablemente el error de test para ning√∫n valor del par√°metro. Podemos ver que a medida que el par√°metro aumenta, hay un momento en que los atributos necesarios para realizar una buena predicci√≥n se eliminan, generando que el error de ambos conjuntos aumente hasta un punto desde donde pareciera constante.

```python
MSE = lambda y, yhat: np.mean(np.power(y - yhat, 2))
Xm = Xtrain.as_matrix()
ym = ytrain.as_matrix()
from sklearn import cross_validation
k_fold = cross_validation.KFold(len(Xm), 7)
best_cv_mse = float("inf")
alphas_ = np.logspace(2, -3, base=10)
for a in alphas_:
    model = Ridge(fit_intercept=True)
    mse_list_k10 = []
    model.set_params(alpha=a)
    for train, vald in k_fold:
        cross_scaler = StandardScaler().fit(Xm[train])
        df_cross_train = pd.DataFrame(
            cross_scaler.transform(Xm[train]),
            columns=Xtrain.columns)
        df_cross_test = pd.DataFrame(
            cross_scaler.transform(Xm[vald]),
            columns=Xtrain.columns)
        mse_list_k10.append(MSE(
            model.fit(df_cross_train,
                      ym[train]).predict(df_cross_test),
            ym[vald]))
    if np.mean(mse_list_k10) < best_cv_mse:
        best_cv_mse = np.mean(mse_list_k10)
        best_alpha = a
        print("BEST PARAMETER=%f, MSE(CV)=%f" % (best_alpha, best_cv_mse))
```

El mejor par√°metro alpha es cercano a 20. Es importante el haber utilizado cross-validation para determinar este par√°metro y no directamente obtenerlo a partir del test, ya que √©sto podr√≠a producir overfitting y nuestro modelo realmente no estar√≠a generalizando. El objetivo de machine learning es aprender modelos que generalicen bien, no que minimicen funciones objetivo sobre las cuales se entrenan: para esto est√° la interpolaci√≥n. En un setting real, el test set se obtiene luego de que el modelo ha sido entrenado, por eso se paga tan bien dir√≠a el profe Carlitos.
